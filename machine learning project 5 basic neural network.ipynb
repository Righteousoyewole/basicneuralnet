```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns= breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame['labels'] = breast_cancer_dataset.target
```

```python
print(data_frame['labels'])
```

```python
data_frame.info()
```

```python
data_frame.isnull().sum()
```

```python
data_frame.groupby('labels').mean()
```

```python
X = data_frame.drop(columns='labels', axis=1)
Y = data_frame['labels']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                         keras.layers.Dense(20, activation='relu'),
                         keras.layers.Dense(2, activation='sigmoid')])
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics='accuracy')
```

```python
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['training_data','validation_data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['training_data','validation_data'],loc='upper right')
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
